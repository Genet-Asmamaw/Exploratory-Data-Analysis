```python
#import the necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns

#load the iris dataset
df = pd.read_csv("/content/Iris.csv")

#print the fisrt five rows of he dataset
df.head()
```

1. calculating mean

```python
#calculate the statistical value using numpy library
#calculate mean
Mean_value_of_Id = df['Id'].mean()
Mean_value_of_SepalLengthCm = df['SepalLengthCm'].mean()
Mean_value_of_SepalWidthCm = df['SepalWidthCm'].mean()
Mean_value_of_PetalLengthCm = df['PetalLengthCm'].mean()
Mean_value_of_PetalWidthCm = df['PetalWidthCm'].mean()

print("The mean value of id: ", Mean_value_of_Id)
print("The mean value of SepalLengthCm: ", Mean_value_of_SepalLengthCm)
print("The mean value of SepalLengthCm: ", Mean_value_of_SepalWidthCm)
print("The mean value of SepalLengthCm: ", Mean_value_of_PetalLengthCm)
print("The mean value of SepalLengthCm: ", Mean_value_of_PetalWidthCm)
```

2. Calculating median


```python
#calculate median
Median_value_of_Id = df['Id'].median()
Median_value_of_SepalLengthCm = df['SepalLengthCm'].median()
Median_value_of_SepalWidthCm = df['SepalWidthCm'].median()
Median_value_of_PetalLengthCm = df['PetalLengthCm'].median()
Median_value_of_PetalWidthCm = df['PetalWidthCm'].median()

print("The median value of id: ", Median_value_of_Id)
print("The median value of SepalLengthCm: ", Median_value_of_SepalLengthCm)
print("The median value of SepalLengthCm: ", Median_value_of_SepalWidthCm)
print("The median value of SepalLengthCm: ", Median_value_of_PetalLengthCm)
print("The median value of SepalLengthCm: ", Median_value_of_PetalWidthCm)
```

3. Calculating mode

```python
#calculate mode
Mode_value_of_Id = df['Id'].mode()
Mode_value_of_SepalLengthCm = df['SepalLengthCm'].mode()
Mode_value_of_SepalWidthCm = df['SepalWidthCm'].mode()
Mode_value_of_PetalLengthCm = df['PetalLengthCm'].mode()
Mode_value_of_PetalWidthCm = df['PetalWidthCm'].mode()

print("The mode value of id: ", Median_value_of_Id)
print()
print("The mode value of SepalLengthCm: ", Mode_value_of_SepalLengthCm)
print()
print("The mode value of SepalLengthCm: ", Mode_value_of_SepalWidthCm)
print()
print("The mode value of SepalLengthCm: ", Mode_value_of_PetalLengthCm)
print()
print("The mode value of SepalLengthCm: ", Mode_value_of_PetalWidthCm)
```

4. Calculate standard deviation


```python
#Calculate standard deviation
Std_value_of_Id = df['Id'].std()
Std_value_of_SepalLengthCm = df['SepalLengthCm'].std()
Std_value_of_SepalWidthCm = df['SepalWidthCm'].std()
Std_value_of_PetalLengthCm = df['PetalLengthCm'].std()
Std_value_of_PetalWidthCm = df['PetalWidthCm'].std()

print("The std value of id: ", Median_value_of_Id)
print("The std value of SepalLengthCm: ", Std_value_of_SepalLengthCm)
print("The std value of SepalLengthCm: ", Std_value_of_SepalWidthCm)
print("The std value of SepalLengthCm: ", Std_value_of_PetalLengthCm)
print("The std value of SepalLengthCm: ", Std_value_of_PetalWidthCm)
```

5. Calculate Variance

```python
#Calculate variance
Var_value_of_Id = df['Id'].var()
Var_value_of_SepalLengthCm = df['SepalLengthCm'].var()
Var_value_of_SepalWidthCm = df['SepalWidthCm'].var()
Var_value_of_PetalLengthCm = df['PetalLengthCm'].var()
Var_value_of_PetalWidthCm = df['PetalWidthCm'].var()

print("The variance value of id: ", Median_value_of_Id)
print("The variance value of SepalLengthCm: ", Var_value_of_SepalLengthCm)
print("The variance value of SepalLengthCm: ", Var_value_of_SepalWidthCm)
print("The variance value of SepalLengthCm: ", Var_value_of_PetalLengthCm)
print("The variance value of SepalLengthCm: ", Var_value_of_PetalWidthCm)
```
