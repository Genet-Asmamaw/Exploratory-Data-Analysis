1. import the necessary libraries


```python
#import the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from pandas import Series, DataFrame

#load the tips dataset
df = pd.read_csv("/content/tips.csv")
df.head()
```

5. Visualize using bar graph

```python
a=pd.DataFrame(df['day'].value_counts())
a.reset_index(inplace=True)
```

```python
plt.bar(a['index'],a['day'])
```

```python
a.plot(kind='bar',x='index',y='day',colormap='icefire')
plt.xticks(rotation=0)
plt.show()
```

6. Visualize using pie chart

```python
plt.pie(a['day'],labels=a['index'],autopct='%1.2f',explode=[0.2,0,0,0])
plt.show()
```

```python
a.plot(kind='pie',y='day',labels=a['index'],autopct='%1.2f')
plt.show()
```

```python
sns.displot(df['total_bill'])
```

```python
a=df['total_bill']
mean=a.mean()
median=np.median(a)
mode=a.mode()
```

```python
sns.distplot(a,hist=False)
plt.axvline(mean,color='r',label='mean')
plt.axvline(median,color='b',label='median')
plt.axvline(mode[0],color='g',label='mode')
plt.legend()
plt.show()
```

```python
a.plot(kind='kde')
plt.axvline(mean,color='r',label='mean')
plt.axvline(median,color='b',label='median')
plt.axvline(mode[0],color='g',label='mode')
plt.legend
```

```python
a.plot(kind='density')
plt.axvline(mean,color='r',label='mean')
plt.axvline(median,color='b',label='median')
plt.axvline(mode[0],color='g',label='mode')
plt.legend
```

```python
plt.boxplot(a)
plt.text(0.85,13,s='Q1',size=13)
plt.text(0.85,17,s='Q2',size=13)
plt.text(0.85,23,s='Q3',size=13)
plt.text(1.1,16,s='IQR',rotation=90,size=20)
plt.show()
```

```python
x1=[1,2,3,4]
y1=[5,10,3,20]
x2=[10,15,12,11]
plt.plot(x1,y1,linestyle='-.',marker='^',markersize=12,color='green',label='line 1')
plt.plot(x1,x2,linestyle='--',marker='o',markersize=12,color='red',label='line 2')
plt.title('Two Line Graphs')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.grid()
plt.show()
```
